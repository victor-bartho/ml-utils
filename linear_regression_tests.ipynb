```python
import pandas as pd
import numpy as np
```

Ideal equations taken from: https://www.graphpad.com/quickcalcs/linear1/

Ideal Equation for 10 records dataset: Y = 2*X

Ideal Equation for 50 records dataset: Y = 4.071*X + 2.369

Ideal Equation for Tanu N Prabhu dataset: Y = 0.6400*X + 5.889  
    - Tanu dataset taken from: https://www.kaggle.com/datasets/tanuprabhu/linear-regression-dataset/data

```python
#df = pd.read_csv('datasets/simpleregression_gpt_10records.csv')
df = pd.read_csv('datasets/Tanu_N_Prabhu_Linear_Regression.csv')
#df = pd.read_csv('datasets/simpleregression_gpt_50records.csv')
print(df)
```

```python
training_data = df['X'].to_numpy()
print(training_data)
```

```python
target_values = df['Y'].to_numpy()
print(target_values)
```

#### Normalization os values to avoid scalar overflow

```python
training_data_mean = np.mean(training_data)
training_data_std = np.std(training_data)
training_data_normalized = (training_data - training_data_mean) / training_data_std
print(training_data_normalized)
```

```python
target_values_mean = np.mean(target_values)
target_values_std = np.std(target_values)
target_values_normalized = (target_values - target_values_mean)/target_values_std
print(target_values_normalized)
```

```python
from linear_regression_utils import calculate_gradient_descent
b_initial = 0
w_initial = 0
alpha = np.float64(5.0e-2)
iter_number = 100000
w_final, b_final = calculate_gradient_descent(training_data_normalized, target_values_normalized, w_initial, b_initial, iter_number, alpha)
print(w_final)
print(b_final)
```

#### Converting values back to original scales ("denormalization")

```python
w_original = w_final *(target_values_std/training_data_std)
b_original = target_values_mean - (w_original * training_data_mean) - (b_final * target_values_std)
print(w_original)
print(b_original)
```
